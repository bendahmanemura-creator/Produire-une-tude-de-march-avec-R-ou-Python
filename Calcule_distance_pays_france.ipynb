## Calculer la distance entre la france et d'autre pays 

### Importation des librairies

```python
df_pays = pd.read_csv("Countries & Territories Taxonomy MVP - C&T Taxonomy with HXL Tags.csv")
```

```python
print(df_pays.columns.tolist())
```

```python
# Coordonnées approximatives de la France (Paris)
france_lat, france_lon = 48.8566, 2.3522

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi/2)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(delta_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

# Enlever les lignes sans coordonnées
df_pays['Latitude'] = pd.to_numeric(df_pays['Latitude'], errors='coerce').astype('float')
df_pays['Longitude'] = pd.to_numeric(df_pays['Longitude'], errors='coerce').astype('float')
coords = df_pays.dropna(subset=['Latitude', 'Longitude'])

# Calcul distance
coords['Distance_km'] = coords.apply(lambda x: haversine(france_lat, france_lon, x['Latitude'], x['Longitude']), axis=1)

# Extraire nom pays et distance
result = coords[['Preferred Term', 'Distance_km']].rename(columns={'Preferred Term': 'Pays'})

# Sauvegarder en CSV
result.to_csv('distance_vers_france.csv', index=False)
```
